```python
import numpy as np
import pandas as pd

# Read in the data

data = pd.read_csv('Historical Weather Data Central Park NY 2017.csv')
```

```python
# Slicing the original data to obtain a desired data frame

data_sod = data[['DATE', 'REPORTTYPE', 'DAILYMaximumDryBulbTemp', 'DAILYMinimumDryBulbTemp', 'DAILYAverageDryBulbTemp', 'DAILYPrecip', 'DAILYSnowfall', 'DAILYSnowDepth', 'DAILYAverageWindSpeed', 'DAILYPeakWindSpeed']]
data_sod = data_sod.loc[data_sod.REPORTTYPE == 'SOD']
data_sod.head(50)
```

__DATA CLEANING__

```python
# Forward fill the weather data of the missing day

temp_row = data_sod.loc[[330]]
temp_row.DATE = '2017-01-11 23:59:00'
temp_row.index = [331]
data_sod = data_sod.append(temp_row)
data_sod = data_sod.sort_index()
```

```python
# Define a function to extract only date info

def get_date_only(df):
    temp_list = df.DATE.split()
    return temp_list[0]
```

```python
# Add DATE_ONLY column to our data frame

date_only_col = data_sod.apply(get_date_only, axis=1)
data_sod['DATE_ONLY'] = date_only_col
data_sod
```

```python
# Drop useless columns

data_sod.drop(columns=['DATE', 'REPORTTYPE'], inplace=True)
```

```python
# Forward fill all missing values

data_sod.isnull().any(axis=0)
data_sod.fillna(method='ffill', inplace=True)
data_sod
```

```python
# Replace all T's with the value 0

data_sod.replace('T', 0, inplace=True)
data_sod
```

```python
# Save the cleaned data frame in a csv file

data_sod.to_csv('Daily NY Weather Cleaned.csv', index=False)
```
